```python
print("Given System of Linear Equations:")
print("1) 2x + 3y = 10")
print("2) 4x + 5y = 20")

A = [
    [2, 3],
    [4, 5]
]

B = [10, 20]

# Determinant of 2x2 matrix
def determinant(A):
    return A[0][0]*A[1][1] - A[0][1]*A[1][0]

# Inverse of 2x2 matrix
def inverse(A):
    det = determinant(A)

    return [
        [ A[1][1]/det, -A[0][1]/det],
        [-A[1][0]/det,  A[0][0]/det]
    ]

# Solve AX = B
def solve(invA, B):
    x = invA[0][0]*B[0] + invA[0][1]*B[1]
    y = invA[1][0]*B[0] + invA[1][1]*B[1]
    return x, y

print("\nCoefficient Matrix A:")
print(A[0])
print(A[1])

detA = determinant(A)
print("\nDeterminant of Matrix A:", detA)

invA = inverse(A)
print("\nInverse of Matrix A:")
print(invA[0])
print(invA[1])

x, y = solve(invA, B)
print("\nSolution of the given equations:")
print("x =", round(x, 2))
print("y =", round(y, 2))
```

```python
import numpy as np

print("Given System of Linear Equations:")
print("1) 2x + 3y = 10")
print("2) 4x + 5y = 20")

A = np.array([
    [2, 3],
    [4, 5]
])

B = np.array([10, 20])

print("\nCoefficient Matrix A:")
print(A)

print("\nConstant Matrix B:")
print(B)

detA = np.linalg.det(A)
print("\nDeterminant of Matrix A:", round(detA, 2))

invA = np.linalg.inv(A)
print("\nInverse of Matrix A:")
print(np.round(invA, 2))

solution = np.linalg.solve(A, B)
print("\nSolution of the given equations:")
print("x =", round(solution[0], 2))
print("y =", round(solution[1], 2))
```
