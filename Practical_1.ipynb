###CALCULATE THE FACTORIAL OF A NUMBER TAKEN AS INPUT FROM USER
####PROMPT:WAP to compute factorial of a given number


```python
def factorial(n):
    if n == 0:
        return 1
    else:
        res = 1
        for i in range(1, n + 1):
            res *= i
        return res

# Example usage:
num = int(input("Enter a non-negative integer: "))
print(f"The factorial of {num} is {factorial(num)}")
```

####Display table for a given number
#####PROMPT:WAP to display table for a given number

```python
def display_multiplication_table(number):
    print(f"Multiplication Table for {number}:")
    for i in range(1, 11):
        print(f"{number} x {i} = {number * i}")

# Get input from the user
num_to_table = int(input("Enter a number to display its multiplication table: "))

# Display the multiplication table
display_multiplication_table(num_to_table)
```

###CHECK WETHER THE NUMBER IS EVEN OR ODD
####PROMPT:WAP to identify the given number is odd or even

```python
def check_odd_even(number):
    if number % 2 == 0:
        return "Even"
    else:
        return "Odd"

# Get input from the user
num_to_check = int(input("Enter an integer to check if it's odd or even: "))

# Check and display the result
if num_to_check == 0:
    print(f"The number {num_to_check} is zero (which is considered an even number).")
else:
    result = check_odd_even(num_to_check)
    print(f"The number {num_to_check} is {result}.")
```

###Identify wether the entered number is palindrome or not.
####Prompt:WAP to identify the entered number is palindrome or not.

```python
def is_palindrome(number):
    # Convert the number to a string to easily reverse it
    num_str = str(number)
    # Compare the original string with its reverse
    if num_str == num_str[::-1]:
        return True
    else:
        return False

# Get input from the user
num_to_check = int(input("Enter an integer to check if it's a palindrome: "))

# Check and display the result
if is_palindrome(num_to_check):
    print(f"The number {num_to_check} is a palindrome.")
else:
    print(f"The number {num_to_check} is not a palindrome.")
```

###Prompt:Develop an Expert system to determine student grade based on marks.

```python
def grade_expert_system(marks):
    """
    Expert system to determine student grade based on marks
    """

    if marks < 0 or marks > 100:
        return "Invalid marks! Please enter a value between 0 and 100."

    # Knowledge base rules
    if marks >= 90:
        return "Grade: A"
    elif marks >= 80:
        return "Grade: B"
    elif marks >= 70:
        return "Grade: C"
    elif marks >= 60:
        return "Grade: D"
    else:
        return "Grade: F"


# User interaction
try:
    student_marks = float(input("Enter student marks (0–100): "))
    result = grade_expert_system(student_marks)
    print(result)
except ValueError:
    print("Please enter a valid numeric value.")
```
