```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Upload file (Colab)
from google.colab import files
uploaded = files.upload()

# Load dataset
df = pd.read_csv("Student_Performance.csv")

# Display first 5 rows
print("First 5 rows of dataset:")
print(df.head())

# -------------------------------
# 1. Filtering Data
# -------------------------------

filtered_data = df[df['overall_score'] > 70]

print("\nFiltered Data (Overall Score > 70):")
print(filtered_data.head())

# -------------------------------
# 2. Grouping & Aggregate Statistics
# -------------------------------

grouped_stats = df.groupby('school_type')['study_hours'].agg([
    'sum', 'count', 'mean', 'median', 'std', 'var', 'min', 'max'
])

print("\nGrouped Statistics (School_Type vs Hours_Studied):")
print(grouped_stats)
```

```python
# -------------------------------
# Interpretation of Insights
# -------------------------------

print("\n--- Interpretation of Insights ---")
# 1. High performers
high_performers = df[df['overall_score'] > 70]
print(f"Number of high-performing students: {len(high_performers)}")
# 2. Average study hours by school type
avg_hours = grouped_stats['mean']
print("\nAverage study hours by School Type:")
print(avg_hours)
# 3. Compare which school studies more
max_school = avg_hours.idxmax()
print(f"\nSchool type with highest average study hours: {max_school}")
# 4. Variability insight
std_dev = grouped_stats['std']
print("\nVariation in study hours (Std Dev):",std_dev)
# 5. Range insight
min_hours = grouped_stats['min']
max_hours = grouped_stats['max']

print("\nMinimum study hours:",min_hours)
print("\nMaximum study hours:",max_hours)
```
