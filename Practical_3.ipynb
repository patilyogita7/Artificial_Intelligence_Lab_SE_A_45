```python
# Inverse of 3x3 matrix without numpy

def determinant(m):
  return (
m[0][0]*(m[1][1]*m[2][2] - m[1][2]*m[2][1])
- m[0][1]*(m[1][0]*m[2][2] - m[1][2]*m[2][0])
+ m[0][2]*(m[1][0]*m[2][1] - m[1][1]*m[2][0])
)

def inverse_3x3(m):
    det = determinant(m)

    if det == 0:
        print("Matrix has no inverse")
        return None

    # Cofactor matrix
    cof = [
        [
            (m[1][1]*m[2][2] - m[1][2]*m[2][1]),
            -(m[1][0]*m[2][2] - m[1][2]*m[2][0]),
            (m[1][0]*m[2][1] - m[1][1]*m[2][0])
        ],
        [
            -(m[0][1]*m[2][2] - m[0][2]*m[2][1]),
            (m[0][0]*m[2][2] - m[0][2]*m[2][0]),
            -(m[0][0]*m[2][1] - m[0][1]*m[2][0])
        ],
        [
            (m[0][1]*m[1][2] - m[0][2]*m[1][1]),
            -(m[0][0]*m[1][2] - m[0][2]*m[1][0]),
            (m[0][0]*m[1][1] - m[0][1]*m[1][0])
        ]
    ]

    # Transpose of cofactor (adjoint)
    adj = [[cof[j][i] for j in range(3)] for i in range(3)]

    # Divide by determinant
    inv = [[adj[i][j]/det for j in range(3)] for i in range(3)]
    return inv

# Example matrix

matrix = [
[1, 2, 3],
[0, 1, 4],
[5, 6, 0]
]

result = inverse_3x3(matrix)

print("Inverse matrix:")
for row in result:
    print(row)
```

```python
import numpy as np

# Create a 3x3 matrix
matrix_np = np.array([
    [1, 2, 3],
    [0, 1, 4],
    [5, 6, 0]
])

print("Original 3x3 matrix:")
print(matrix_np)

# Compute the determinant
determinant_np = np.linalg.det(matrix_np)

print(f"\nDeterminant of the matrix: {determinant_np}")
```

```python
# Compute the inverse of the matrix
inverse_matrix_np = np.linalg.inv(matrix_np)

print("\nInverse of the matrix:")
print(inverse_matrix_np)
```

```python
import numpy as np

# Define the coefficient matrix A
A = np.array([
    [2, 3],
    [4, 5]
])

# Define the constant vector B
B = np.array([10, 20])

print("Coefficient Matrix A:")
print(A)
print("\nConstant Vector B:")
print(B)

# Solve the system of linear equations Ax = B
x = np.linalg.solve(A, B)

print("\nSolution (x, y):")
print(x)
```
