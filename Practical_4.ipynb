```python
# Experiment 04: Probability & Statistics

import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# 1. Create Normally Distributed Data

avg = 60          # Mean value
std = 8           # Standard deviation
n = 800           # Sample size

values = np.random.normal(avg, std, n)

print("Sample Data (First 10 values):")
print(np.round(values[:10], 2))


# 2. Statistical Measures

avg_value = np.mean(values)
mid_value = np.median(values)
var_value = np.var(values)
std_value = np.std(values)

print("\n--- Statistical Results ---")
print(f"Average (Mean): {round(avg_value, 2)}")
print(f"Median: {round(mid_value, 2)}")
print(f"Variance: {round(var_value, 2)}")
print(f"Standard Deviation: {round(std_value, 2)}")

# 3. Visualization of Distribution

plt.hist(values, bins=25, density=True, alpha=0.5)

left, right = plt.xlim()
x_axis = np.linspace(left, right, 120)
pdf_curve = norm.pdf(x_axis, avg, std)

plt.plot(x_axis, pdf_curve, linewidth=2)
plt.title("Gaussian Distribution Curve")
plt.xlabel("Data Values")
plt.ylabel("Density")
plt.show()

# 4. Probability Simulation (Coin Toss)

trials = 1200
outcomes = np.random.choice(['Head', 'Tail'], size=trials)

count_head = np.count_nonzero(outcomes == 'Head')
count_tail = np.count_nonzero(outcomes == 'Tail')

print("\n--- Coin Toss Experiment ---")
print("Total Trials:", trials)
print("Heads Count:", count_head)
print("Tails Count:", count_tail)
print("P(Head):", round(count_head / trials, 3))
print("P(Tail):", round(count_tail / trials, 3))
```
