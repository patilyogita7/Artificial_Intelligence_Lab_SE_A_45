```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm
```

```python
# 1. Generate a normal distribution
mu_input = input("Enter the mean (mu) for the normal distribution (e.g., 0): ")
sigma_input = input("Enter the standard deviation (sigma) for the normal distribution (e.g., 0.1): ")

# Convert inputs to float
mu = float(mu_input)
sigma = float(sigma_input)

s = np.random.normal(mu, sigma, 1000)
print("First 10 generated values")
print(s[:10])
# 2. Plot its histogram
plt.figure(figsize=(8, 6))
count, bins, ignored = plt.hist(s, 30, density=True)

# Plot the probability density function (PDF) for comparison
plt.plot(bins, 1/(sigma * np.sqrt(2 * np.pi)) * np.exp( - (bins - mu)**2 / (2 * sigma**2) ),
         linewidth=2, color='r')
plt.title('Histogram of a Normal Distribution')
plt.xlabel('Value')
plt.ylabel('Probability Density')
plt.grid(True)
plt.show()
```

```python
data_input_str = input("Please enter a series of numbers, separated by commas (e.g., '1,2,3,4,5'): ")
print(f"You entered: {data_input_str}")
```

```python
mean_data = np.mean(data_array)
median_data = np.median(data_array)
variance_data = np.var(data_array)
std_dev_data = np.std(data_array)

print(f"Mean of data_array: {mean_data}")
print(f"Median of data_array: {median_data}")
print(f"Variance of data_array: {variance_data}")
print(f"Standard Deviation of data_array: {std_dev_data}")
```

```python
num_tosses = 20
coin_outcomes = np.random.choice(['Heads', 'Tails'], size=num_tosses)

print(f"Generated {num_tosses} coin toss outcomes.")
print("First 10 outcomes:", coin_outcomes[:10])
print("Type of coin_outcomes:", type(coin_outcomes))
```

```python
outcomes, counts = np.unique(coin_outcomes, return_counts=True)

heads_count = 0
tails_count = 0

for outcome, count in zip(outcomes, counts):
    if outcome == 'Heads':
        heads_count = count
    elif outcome == 'Tails':
        tails_count = count

print(f"Number of Heads: {heads_count}")
print(f"Number of Tails: {tails_count}")
```

```python
probability_heads = heads_count / num_tosses
probability_tails = tails_count / num_tosses

print(f"Probability of Heads: {probability_heads}")
print(f"Probability of Tails: {probability_tails}")
```

```python
# Import required libraries
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# 1. Generate Normal Distribution Data

mu = 50        # Mean
sigma = 10     # Standard deviation
size = 1000    # Number of samples

# Generate random data
data = np.random.normal(mu, sigma, size)

print("First 10 Generated Values:")
print(data[:10])   # Show only first 10 values

# 2. Descriptive Statistics

mean = np.mean(data)
median = np.median(data)
variance = np.var(data)
std_dev = np.std(data)

print("Descriptive Statistics:")
print("Mean:", mean)
print("Median:", median)
print("Variance:", variance)
print("Standard Deviation:", std_dev)

# 3. Plot Normal Distribution

plt.hist(data, bins=30, density=True, alpha=0.6, color='g')
xmin, xmax = plt.xlim()
x = np.linspace(xmin, xmax, 100)
p = norm.pdf(x, mu, sigma)

plt.plot(x, p, 'k', linewidth=2)
plt.title("Normal Distribution (Mean = 50, Std Dev = 10)")
plt.xlabel("Values")
plt.ylabel("Probability Density")
plt.show()

# 4. Coin Toss Simulation

tosses = 1000
coin = np.random.choice(['H', 'T'], size=tosses)

heads = np.sum(coin == 'H')
tails = np.sum(coin == 'T')

print("\nCoin Toss Simulation:")
print("Total Tosses:", tosses)
print("Heads:", heads)
print("Tails:", tails)
print("Probability of Heads:", heads/tosses)
print("Probability of Tails:", tails/tosses)
```
